```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'],utc=True)
df['Formatted Date']
```

```python
df = df.set_index('Formatted Date')
df
```

```python
data_columns =['Apparent Temperature (C)', 'Humidity']
df_month_mean = df[data_columns].resample('MS').mean()
df_month_mean.tail()
```

```python
plt.figure(figsize =(20,10))
sns.lineplot(data = df_month_mean,marker="o")
plt.title("Variation in Temperature and humidity with repect to time ")
```

```python
df_month = df_month_mean[df_month_mean.index.month==1]
df_month
```
